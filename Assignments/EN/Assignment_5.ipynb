#### 1. Count words in a text

a. Output a list of words in the file along with their frequency counts (ignoring case).   
b. Count how many unique words there are (ignoring case).    
c. Check how common are all different sequences of vowels (e.g. the sequences "ieu" or just "e" in "lieutenant")?

```python
#a)
import string
import re
cleaned_text=text.lower().translate(str.maketrans('', '', string.punctuation)).replace('\n', ' ')
cleaned_text=re.sub(r'\d+', '', cleaned_text)
words=cleaned_text.strip().split()

count=dict()
for word in words:
    if  word not in count:
           count[word]=1
    else:
        count[word]+=1

counted_list = [(key, value) for key, value in count.items()]
counted_list
```

```python
#b)
unique=set(words)
len(unique)
```

```python
# c) 
reg="[aeiouy]+"
vocals=re.findall(reg,cleaned_text)
vocal_count=dict()

for v in vocals:
    if v not in vocal_count:
        vocal_count[v]=1
    else:
        vocal_count[v]=vocal_count[v]+1

print(vocal_count)
```

#### 4. Compute ngrams  

a. Find the 10 most common bigrams  
b. Find the 10 most common trigrams  

```python
# a)
bigrams=[]

for w in range(len(words)-1):
	bigrams.append((words[w], words[w+1]))
	
freq=dict()

for b in bigrams:
    if b not in freq:
        freq[b]=1
    else:
        freq[b]+=1
        
top10 = sorted(freq.items(), key=lambda x:x[1], reverse=True)[:10]
top10
```

```python
# b) 
trigrams=[]
for w in range(len(words)-2):
	trigrams.append((words[w], words[w+1], words[w+2]))
	
t_freq=dict()

for t in trigrams:
    if t not in t_freq:
        t_freq[t]=1
    else:
        t_freq[t]+=1
        
top10t = sorted(t_freq.items(), key=lambda x:x[1], reverse=True)[:10]
top10t
```
